## Time taken for optimization with different number of steps

```python
import matplotlib.pyplot as plt
import pandas as pd
multi_op = pd.read_csv('data/multiple_optimization.csv')
multi_op['configuration'] = multi_op['Filename'].str.split('-').str[0]
multi_op.drop(columns = ['Filename'], inplace = True)
multi_op
op = multi_op.melt(id_vars = ['configuration'], var_name = 'steps', value_name = 'time_taken', ignore_index=True)
op = op.groupby(['configuration', 'steps'])['time_taken'].agg(['mean', 'median', 'min', 'max', 'std']).reset_index()

op['steps'] = op['steps'].astype(int)
op = op.sort_values(['configuration', 'steps'])

# plt.plot(op)
# plt.show()
plt.figure(figsize=(10, 6))
label_names = ['54 atoms - evenly distributed', 
               '16 atoms - unevenly distributed',
               '250 atoms - unevenly distributed',
               '128 atoms - evenly distributed',
               '16 atoms - evenly distributed',
               '54 atoms - unevenly distributed',
               '250 atoms - evenly distributed',
               '128 atoms - evenly distributed']

i = -1
for config in op['configuration'].unique():
    subset = op[op['configuration'] == config]
    #plt.plot(subset['steps'], subset['median'], marker='.', label=f'Config {config}')
    plt.errorbar(subset['steps'], subset['median'], subset['std'], fmt='o-', label=f'{label_names[i]}')
    i+=1
plt.title('Median Time Taken per Step for Each Configuration')
plt.xlabel('Steps')
plt.ylabel('Median Time Taken (s)')
plt.legend(title='Configuration')
plt.grid(True)
plt.show()

# large error bar for the pink line because anomaly was present where one configuration converged very quickly
```

## Plotting how fmax & energy from log files changes wrt number of optimization steps (produced from optimizations)

```python
import os
import pandas as pd
folder = 'data/logfiles'
for files in os.listdir(folder):
    if files.endswith('.log'):
        filename = os.path.join(folder, files)

        max_df = pd.read_csv(filename, sep='\s+')
        max_df = max_df[max_df['Step'] <= 5000]

        import matplotlib.pyplot as plt

        # Create a figure with 1 row and 2 columns
        fig, axes = plt.subplots(1, 2, figsize=(11, 5))  # (1 row, 2 columns)

        # First plot: Step vs Energy
        axes[0].plot(max_df['Step'], max_df['Energy'], linestyle='-', color='b')
        axes[0].set_title(f'Potential Energy vs Step for {files[:-9]}')
        axes[0].set_xlabel('Step')
        axes[0].set_ylabel('Potential Energy (eV)')
        axes[0].grid(True)

        # Second plot: Step vs fmax (if available)
        axes[1].plot(max_df['Step'], max_df['fmax'], linestyle='-', color='r')
        axes[1].set_title(f'fmax vs Step for {files[:-9]}')
        axes[1].set_xlabel('Step')
        axes[1].set_ylabel('fmax (eV/Å)')
        axes[1].grid(True)

        # Adjust layout and show the plots
        plt.tight_layout()
        plt.show()
    else:
        continue
```

## Extracting structures from logfiles

```python
from ase.io import read, write
from ase.io import Trajectory
import os

folder = 'data/logfiles'
target = []
for index in range(0, 520, 20):
    target.append(index)
    
for filename in os.listdir(folder):
    if filename.endswith('.traj'):
        files = os.path.join(folder, filename)
        for n in target:
            os.makedirs('data/traj_structure', exist_ok=True)
            output = f'data/traj_structure/{filename}_{n}.vasp'
            n_atom = read(files, index = n)
            write(output, n_atom, format = 'vasp', direct = True)
    else:
        continue
```

## How Pugh ratio changes with optimization steps

```python
import pandas as pd
pred = pd.read_csv('data/output_structure_predictions.csv')
pred['steps'] = pred['File Name'].str.split('_').str[1]
pred['steps'] = pred['steps'].str.split('.').str[0]
pred['steps'] = pred['steps'].astype(int)
pred['configuration'] = pred['File Name'].str.split('-').str[0]
pred['arrangement'] = pred['File Name'].str.split('-').str[1]
pred['arrangement'] = pred['arrangement'].str.split('_').str[0].str[0]
pred.drop(columns = 'File Name', inplace = True)
pred.sort_values(['configuration', 'arrangement', 'steps'], inplace = True)
pred = pred[['configuration', 'arrangement', 'steps', 'pugh_ratio']]
# pred.groupby(['configuration', 'arrangement'])

pred_sorted = pred.sort_values(by=['configuration', 'arrangement'])

arrangement_sample = pred_sorted.groupby('configuration').apply(lambda x: x.iloc[0]).reset_index(drop=True)

sampled = pd.merge(pred, arrangement_sample[['configuration', 'arrangement']], on=['configuration', 'arrangement'], how='inner')
sampled

import matplotlib.pyplot as plt
max_steps = pred_sorted[pred_sorted['steps'] == 500]

variance_per_step = max_steps.groupby(['configuration'])['pugh_ratio'].std().reset_index().rename(columns = {'pugh_ratio': 'variance'})
variance_per_step
# 16, 54, 128, 250

label_names = ['54 atoms - evenly distributed', 
               '16 atoms - unevenly distributed',
               '250 atoms - unevenly distributed',
               '128 atoms - evenly distributed',
               '16 atoms - evenly distributed',
               '54 atoms - unevenly distributed',
               '250 atoms - evenly distributed',
               '128 atoms - unevenly distributed']
i = 0
for config in pred['configuration'].unique():
    # Filter the data for the current configuration
    config_data = pred[pred['configuration'] == config]
    plt.figure(figsize=(8, 5))

    # Plot the data for the current configuration
    for arrangement in pred['arrangement'].unique():
        fig = config_data[config_data['arrangement'] == arrangement]
        plt.plot(fig['steps'], fig['pugh_ratio'], marker='.', label=arrangement)

    variance_value = variance_per_step.loc[variance_per_step['configuration'] == config, 'variance'].values[0]

    # Add variance annotation inside the plot
    plt.text(0.05, 0.95, f'std at 200 steps: {variance_value:.8f}', 
             transform=plt.gca().transAxes, fontsize=15, color='red', 
             verticalalignment='top', bbox=dict(facecolor='white', alpha=0.7))


    plt.title(f'Pugh ratio vs Steps for Configuration with {label_names[i]}', fontsize = 15)
    plt.xlabel('Steps', fontsize = 15)
    plt.ylabel('Pugh Ratio', fontsize = 15)
    plt.ylim(0.15, 0.35)
    #plt.legend(title='Arrangement', bbox_to_anchor=(1.05, 1), loc='upper left')

    plt.tight_layout()
    plt.show()
    i+=1

```

## Parity plot of predicted intermetallic properties

### Bulk and Shear Modulus
- Intermetallic compounds pugh ratio are calculated from bulk and shear modulus models predictions

```python
# Bulk and shear model
bs_pred = pd.read_csv('data/intermetallics_2and3_predictions.csv')
theoretical = pd.read_csv('data/intermetallics_2and3_theoretical.csv')

bs_pred['material'] = bs_pred['File Name'].str.split('_').str[0]
bs_pred['material_id'] = bs_pred['File Name'].str.split('_').str[1]
bs_pred['symmetry'] = bs_pred['File Name'].str.split('_').str[2].str[:-5]
bs_pred = bs_pred[['material', 'material_id', 'symmetry', 'shear (GPa)', 'bulk (GPa)', 'pugh_ratio']]
bs_pred.rename(columns = {'shear (GPa)': 'shear_modulus',
                        'bulk (GPa)': 'bulk_modulus'}, inplace=True)

compare_df = theoretical.merge(bs_pred, how = 'inner', on = ['material', 'material_id', 'symmetry'], suffixes=('_theoretical', '_predicted'))
compare_df[['pugh_ratio_theoretical', 'pugh_ratio_predicted']]
compare_df = compare_df.dropna()
compare_df = compare_df[compare_df['pugh_ratio_theoretical'] >= 0]
compare_df
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import matplotlib.pyplot as plt
mse = mean_squared_error(compare_df['pugh_ratio_theoretical'], compare_df['pugh_ratio_predicted'])
mae = mean_absolute_error(compare_df['pugh_ratio_theoretical'], compare_df['pugh_ratio_predicted'])
r_squared = r2_score(compare_df['pugh_ratio_theoretical'], compare_df['pugh_ratio_predicted'])
print('mse:', mse)
print('mae:', mae)
print('r2_score', r_squared)

plt.figure(figsize=(8, 6))
plt.scatter(compare_df['pugh_ratio_theoretical'], compare_df['pugh_ratio_predicted'], alpha=0.7, label='Predicted vs. Theoretical')

# Identity line (y = x)
min_val = min(compare_df['pugh_ratio_theoretical'].min(), compare_df['pugh_ratio_predicted'].min())
max_val = max(compare_df['pugh_ratio_theoretical'].max(), compare_df['pugh_ratio_predicted'].max())
plt.plot([min_val, max_val], [min_val, max_val], 'r--', label='Perfect Prediction (y=x)')

plt.xlabel('Pugh Ratio Theoretical')
plt.ylabel('Pugh Ratio Predicted')
plt.title('Comparison of Predicted vs. Theoretical Pugh Ratio')
plt.text(0.05, 0.95, f'MSE: {mse:.4f}', transform=plt.gca().transAxes, fontsize=12, verticalalignment='top')
plt.text(0.05, 0.90, f'MAE: {mae:.4f}', transform=plt.gca().transAxes, fontsize=12, verticalalignment='top')
plt.text(0.05, 0.85, f'R2 score: {r_squared:.4f}', transform=plt.gca().transAxes, fontsize=12, verticalalignment='top')

plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
# ductile is 0, brittle is 1
compare_df['theoretical_cat'] = compare_df.apply(lambda x: '1' if x['pugh_ratio_theoretical'] >= 0.57 else '0', axis = 1)
compare_df['predicted_cat'] = compare_df.apply(lambda x: '1' if x['pugh_ratio_predicted'] >= 0.57 else '0', axis=1)

conf_mat = confusion_matrix(compare_df['theoretical_cat'], compare_df['predicted_cat'])
print(f'TN: {conf_mat[0,0]}')
print(f'FP: {conf_mat[0,1]}')
print(f'FN: {conf_mat[1,0]}')
print(f'TP: {conf_mat[1,1]}')

accuracy = accuracy_score(compare_df['theoretical_cat'], compare_df['predicted_cat'])

precision = precision_score(compare_df['theoretical_cat'], compare_df['predicted_cat'], pos_label= '1')

recall = recall_score(compare_df['theoretical_cat'], compare_df['predicted_cat'], pos_label= '1')

f1 = f1_score(compare_df['theoretical_cat'], compare_df['predicted_cat'], pos_label= '1')

print(f"Accuracy: {accuracy:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1 Score: {f1:.2f}")
```
